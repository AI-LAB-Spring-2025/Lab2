```python


file = open("genome.txt", "r")

genome_seq = file.read()

file.close()


sequence_list = list(genome_seq)
sequence_list.pop(-1)

length_string = len(genome_seq)
length_list = len(sequence_list)

print("DNA Sequence:", genome_seq)
print("Sequence as List:", sequence_list)
print("Length of String:", length_string)
print("Length of List:", length_list)

```

```python
t = np.linspace(0, 4 * np.pi,length_string) # 4*pi gives about 2 turns
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5,length_string) # z increases linearly to spread out the helix vertically
# Combine the coordinates into a (genome_length x 3) array
coordinates = np.column_stack((x, y, z))

```

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
 

for i, nucleotide in enumerate(sequence_list):
 ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
 color=colors[nucleotide], marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()
```

```python

import pandas as pd



iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title(' Sepal Length vs. Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['sepal length (cm)'], bins=10)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram ')
plt.show()
```
