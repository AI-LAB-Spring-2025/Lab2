```python
import matplotlib.pyplot as plt
A = [12, 15, 14, 13, 16, 18, 19, 15, 14, 20, 17, 14, 15, 40, 45, 50, 62]
B = [12, 17, 15, 13, 19, 20, 21, 18, 17, 16, 15, 14, 16, 15]
fig, axes = plt.subplots(1, 2, figsize=(5, 5))
axes[0].boxplot(A)
axes[0].set_title("Box Plot for Group A")
axes[0].set_ylabel("Measurement Values")
axes[1].boxplot(B)
axes[1].set_title("Box Plot for Group B")
axes[1].set_ylabel("Measurement Values")
fig.suptitle("Comparison of Group A and Group B")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
with open("genome.txt", "r") as file:
    g = list(file.read().strip())
    gl = len(g)
t = np.linspace(0, 4 * np.pi, gl)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, gl)
color_map = {'A': 'red', 'T': 'blue', 'G': 'green', 'C': 'yellow'}
colors = [color_map.get(base, 'black') for base in g]
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=colors, s=50)
ax.set_title("3D Genome ")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
image_path = "/content/pic.jpg"
img = Image.open(image_path)
img_array = np.array(img)
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_array)
plt.title("Original Image")
plt.axis("off")
rotated_img = np.rot90(img_array)
flipped_img = np.fliplr(img_array)
plt.subplot(1, 3, 2)
plt.imshow(rotated_img)
plt.title("Rotated Image")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(flipped_img)
plt.title("Grayscale Image")
plt.axis("off")
plt.show()

```
