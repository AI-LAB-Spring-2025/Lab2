```python
import numpy as np
import matplotlib.pyplot as plt

file = open("genome.txt", "r")
# Read the first line
line = file.readline().strip() # strip newline characters from the line
genome_sequence=[]
for char in line:
  genome_sequence.append(char)

genome_length=len(genome_sequence)

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length) # z increases linearly to
color_map = {
    'A': 'red',
    'T': 'blue',
    'C': 'green',
    'G': 'orange'
}


colors = [color_map[base] for base in genome_sequence if base in color_map]


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(x, y, z, c=colors, s=80)  # s is the size of the points


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Helix of Genome Sequence')


plt.show()


coordinates = np.column_stack((x, y, z))
```

```python
import numpy as np
import requests
from PIL import Image
import matplotlib.pyplot as plt
from io import BytesIO

# Step 1: Download the image from a URL
image_url = 'https://i.pinimg.com/736x/0e/ff/fe/0efffe66441e709e5995a0d64bc4095e.jpg'  # Replace with an actual image URL
response = requests.get(image_url)

# Step 2: Open the image using PIL
image = Image.open(BytesIO(response.content))

# Step 3: Convert the image to a NumPy array
image_array = np.array(image)

# Step 4: Plot the image using Matplotlib
plt.title("original image")
plt.imshow(image_array)
plt.axis('off')  # Hide axes for better visualization
plt.show()

rotated_image=np.rot90(image_array)
plt.title("rotated image")
plt.imshow(rotated_image)
plt.axis('off')  # Hide axes for better visualization
plt.show()

flipped_image=np.fliplr(image_array)
plt.title("flipped image")
plt.imshow(flipped_image)
plt.axis('off')  # Hide axes for better visualization
plt.show()

gray_img = np.dot (image_array[..., :3], [0.299, 0.587, 0.114])
plt.title("greyscale image")
plt.imshow(gray_img, cmap='grey')
plt.axis('off')  # Hide axes for better visualization
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petalwidth) #Accessing the target labels (species)
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 for versicolor, 2 for virginica)

mean_values = np.mean(X, axis=0)
median_values = np.median(X, axis=0)
std_values = np.std(X, axis=0)


min_values = np.min(X, axis=0)
max_values = np.max(X, axis=0)

sepal_features = X[:, :2]

# Print the results
print("Mean Values for each feature:")
print(mean_values)

print("\nMedian Values for each feature:")
print(median_values)

print("\nStandard Deviation for each feature:")
print(std_values)

print("\nMinimum Values for each feature:")
print(min_values)

print("\nMaximum Values for each feature:")
print(max_values)

# scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], color='blue', edgecolor='black', alpha=0.7)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()

# histogram
plt.figure(figsize=(10, 6))
plt.hist(X[:, 0], bins=15, color='green', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# line
plt.figure(figsize=(10, 6))
plt.plot(X[:, 2], X[:, 3], marker='o', color='purple', linestyle='-', alpha=0.7)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid(True)
plt.show()


```
