```python
import matplotlib.pyplot as plt
import numpy as np

file = open('genome.txt', 'r')

list = []
for line in file:
    for char in line:
        list.append(char)
    
list.pop(-1)

fig = plt.figure()
ax = fig.add_subplot(projection ='3d')

t = np.linspace(0, 4*np.pi, len(list))
x= np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, len(list))

coords= np.column_stack((x, y, z))

colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}

for i, nucleotide in enumerate(list):
    ax.scatter(coords[i,0], coords[i,1], coords[i,2], color=colors[nucleotide])
    
plt.show()
    
```
