```python

f=open('genome.txt','r')
file_data=f.read()
file_data=file_data.strip()
f.close()

genome_list=list(file_data)
genome_len=len(file_data)
# x = cos(t), y = sin(t), z = t (or a scaled version of t)
# We want to span a range so that the helix makes a few turns.
t = np.linspace(0, 4 * np.pi, genome_len)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_len)
# Combine the coordinates into a (genome_length x 3) array
coordinates = np.column_stack((x, y, z))

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(x, y, z, color = "green")
plt.title("simple 3D scatter plot")

 # Generate random genome sequence
nucleotide = ["A", "T", "C", "G"]
colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
# Plot each nucleotide with its corresponding color and coordinates
for i, nucleotide in enumerate(genome_list):
 ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],color=colors[nucleotide], marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

try:
    img  = Image.open('image.png')
    img_array=np.array(img)
    rgb_image = img_array[...,:3]
    plt.imshow(rgb_image)
    plt.show()

    rotate_img=np.rot90(img)
    img_array1=np.array(rotate_img)
    rgb_image1 = img_array1[:, :, :3]
    plt.imshow(rgb_image1)
    plt.show()

    flip_img=np.fliplr(img)
    img_array2=np.array(flip_img)
    rgb_image2 = img_array2[:, :, :3]
    plt.imshow(rgb_image2)
    plt.show()

    gray_img = np.dot (img_array[..., :3], [0.299, 0.587, 0.114])
    plt.imshow(gray_img)
    plt.show()

except IOError:
    pass
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('Iris.csv')

print('mean for SepalLengthCm',df['SepalLengthCm'].mean())
print('mean for SepalWidthCm',df['SepalWidthCm'].mean())
print('mean for PetalLengthCm',df['PetalLengthCm'].mean())
print('mean for PetalWidthCm',df['PetalWidthCm'].mean())

print('median for SepalLengthCm',df['SepalLengthCm'].median())
print('median for SepalWidthCm',df['SepalWidthCm'].median())
print('median for PetalLengthCm',df['PetalLengthCm'].median())
print('median for PetalWidthCm',df['PetalWidthCm'].median())

print('std for SepalLengthCm',df['SepalLengthCm'].std())
print('std for SepalWidthCm',df['SepalWidthCm'].std())
print('std for PetalLengthCm',df['PetalLengthCm'].std())
print('std for PetalWidthCm',df['PetalWidthCm'].std())

print('min for SepalLengthCm',df['SepalLengthCm'].min())
print('min for SepalWidthCm',df['SepalWidthCm'].min())
print('min for PetalLengthCm',df['PetalLengthCm'].min())
print('min for PetalWidthCm',df['PetalWidthCm'].min())

print('max for SepalLengthCm',df['SepalLengthCm'].max())
print('max for SepalWidthCm',df['SepalWidthCm'].max())
print('max for PetalLengthCm',df['PetalLengthCm'].max())
print('max for PetalWidthCm',df['PetalWidthCm'].max())

sepal_length=np.array(df['SepalLengthCm'])
sepal_width=np.array(df['SepalWidthCm'])

plt.scatter(sepal_length, sepal_width)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('scattet plot')
plt.show()


plt.hist(sepal_length)
plt.title('histogram')
plt.show()

petal_length=np.array(df['PetalLengthCm'])
petal_width=np.array(df['PetalWidthCm'])

plt.plot(petal_length, linestyle = 'dotted')
plt.title('petal length')
plt.ylabel('petal length')
plt.show()

plt.plot(petal_width, linestyle = 'dotted')
plt.title('petal width')
plt.ylabel('petal width')
plt.show()
```
