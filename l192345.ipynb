```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

with open('genome.txt', 'r') as file:
    genome_sequence = file.read().strip()

genome_length = len(genome_sequence)

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.sin(t)
y = np.cos(t)
z = np.linspace(0, 5, genome_length)
coordinates = np.vstack((x, y, z)).T

color_map = {'A': 'red', 'T': 'green', 'C': 'blue', 'G': 'yellow'}
colors = [color_map.get(nucleotide, 'gray') for nucleotide in genome_sequence]

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(coordinates[:, 0], coordinates[:, 1], coordinates[:, 2], c=colors, s=10)
ax.set_title('3D Helix Structure of Genome Sequence')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img = Image.open('image.jpg')
img_array = np.array(img)

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.imshow(img_array)
plt.title('Original Image')

rotated_img = np.array(img.rotate(90))
plt.subplot(2, 2, 2)
plt.imshow(rotated_img)
plt.title('Rotated 90Â°')

flipped_img = np.array(img.transpose(Image.FLIP_LEFT_RIGHT))
plt.subplot(2, 2, 3)
plt.imshow(flipped_img)
plt.title('Flipped Horizontally')

gray_img = np.array(img.convert('L'))
plt.subplot(2, 2, 4)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

mean_vals = df.mean(axis=0)
median_vals = df.median(axis=0)
std_vals = df.std(axis=0)
min_vals = df.min(axis=0)
max_vals = df.max(axis=0)

print("Mean:", mean_vals[:4])
print("Median:", median_vals[:4])
print("Std Dev:", std_vals[:4])
print("Min:", min_vals[:4])
print("Max:", max_vals[:4])

sepal_features = df[['sepal length (cm)', 'sepal width (cm)']]

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Width')

plt.subplot(1, 3, 2)
plt.hist(df['sepal length (cm)'], bins=20)
plt.xlabel('Sepal Length')
plt.title('Sepal Length Distribution')

sorted_df = df.sort_values(by='petal length (cm)')
plt.subplot(1, 3, 3)
plt.plot(sorted_df['petal length (cm)'], sorted_df['petal width (cm)'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Width')

plt.tight_layout()
plt.show()
```
