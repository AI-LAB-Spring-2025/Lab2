```python
# Open the file in read mode
file = open("genome.txt", "r")
# Read the entire content of the file
genome_sequence = file.read()
genome_list = list(genome_sequence)
print(len(genome_list))
genome_length = len(genome_list)
# Print the content
genome=""
print(genome_sequence)

# We'll use the parametric equations for a helix: 
#   x = cos(t), y = sin(t), z = t (or a scaled version of t) 
# We want to span a range so that the helix makes a few turns. 
t = np.linspace(0, 4 * np.pi, genome_length)  # 4*pi gives about 2 turns 
x = np.cos(t) 
y = np.sin(t) 
z = np.linspace(0, 5, genome_length)  # z increases linearly to spread out the helix vertically 
# Combine the coordinates into a (genome_length x 3) array 
fig = plt.figure(figsize=(10, 7))
ax = plt.axes(projection="3d")
coordinates = np.column_stack((x, y, z))

# Define colors for each molecule
color_map = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'yellow'}

for i,genome_list in enumerate(genome_list):
    ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2], color=color_map[genome_list], marker='o')


# # Creating plot
# ax.scatter3D(x, y, z, c=colors)
# plt.title("3D scatter plot with colored molecules")
# Close the file
file.close()
```

```python
from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt

image = Image.open('flower.jpeg')

# summarize some details about the image
print(image.format)
print(image.size)
print(image.mode)
numpydata = asarray(image)
print(numpydata)

# Display the image using imshow
plt.imshow(numpydata)

# Add Title
plt.title("Matplotlib Plot NumPy Array") 

# Add Axes Labels
plt.xlabel("x axis") 
plt.ylabel("y axis") 

# Display
plt.show()

rot_img=np.rot90(numpydata)
# Display the image using imshow
plt.imshow(rot_img)

# Add Title
plt.title("Matplotlib Plot NumPy Array") 

# Add Axes Labels
plt.xlabel("x axis") 
plt.ylabel("y axis") 

# Display
plt.show()


flip_img=np.fliplr(numpydata)
# Display the image using imshow
plt.imshow(flip_img)

# Add Title
plt.title("Matplotlib Plot NumPy Array") 

# Add Axes Labels
plt.xlabel("x axis") 
plt.ylabel("y axis") 

# Display
plt.show()
# Grayscale conversion formula: Y = 0.299*R + 0.587*G + 0.114*B 
gray_img = np.dot(numpydata[..., :3], [0.299, 0.587, 0.114])
# plt.imshow(gray_img, cmap=plt.get_cmap('gray'))
plt.imshow(gray_img)
# Add Title
plt.title("Matplotlib Plot NumPy Array") 

# Add Axes Labels
plt.xlabel("x axis") 
plt.ylabel("y axis") 

# Display
plt.show()
```

```python
from sklearn.datasets import load_iris  
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
# Load the Iris dataset  
iris = load_iris()  
# Accessing the features (data) using NumPy array  
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petal width) #Accessing the target labels (species)   
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 forversicolor, 2 for  virginica)

print("mean :",np.mean(X, axis=0))
print("median :",np.median(X, axis=0))
print("std :",np.std(X, axis=0))
print("min :",np.min(X, axis=0))
print("max :",np.max(X, axis=0))
# Extract only the sepal length and sepal width as a NumPy
sepal_len=np.array(X[0])
sepal_wid=np.array(X[1])
print("sepal len",sepal_len)
print("sepal wid",sepal_wid)

plt.scatter(sepal_len,sepal_wid)
plt.xlabel("sepal len")
plt.ylabel("sepal wid")
plt.title("sepal len vs sepal wid")
plt.show()

plt.hist(sepal_len)
plt.xlabel("sepal len")
plt.ylabel("frequency")
plt.title("sepal len vs frequency")
plt.show()
plt.plot(X[2],X[3])
plt.xlabel("petal len")
plt.ylabel("petal wid")
plt.title("petal len vs petal wid")
plt.show()
# sns.lineplot(data=iris, x="petal length", y="pental width")
# plt.show()
```
