```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()
X = np.array(iris.data) 
Y = np.array(iris.target)
```

## Column names

```python
column_names = iris.feature_names
print(column_names)
```

## Mean

```python
x_average = np.mean(X)
y_average = np.mean(Y)

print("X average: ", x_average)
print("Y average: ", y_average)
```

## Median

```python
x_median = np.median(X)
y_median = np.median(Y)

print("X median: ", x_median)
print("Y median: ", y_median)
```

## Standard Deviation

```python
X_std = np.std(X)
Y_std = np.std(Y)   

print("X std: ", X_std) 
print("Y std: ", Y_std)
```

```python
sepal_length = X[:,0]
sepal_width = X[:,1] 

print("Sepal length: ", sepal_length)
print("Sepal width: ", sepal_width)
```

## Scatter plot

```python
plt.scatter(sepal_length, sepal_width)
plt.title("Iris dataset")
plt.xlabel(column_names[0])
plt.ylabel(column_names[1])
plt.show()
```

## Histogram

```python
plt.hist(sepal_length)
plt.title("Sepal length histogram")
plt.show()
```

## Line plot

```python
petal_length = X[:,2]
petal_width = X[:,3]

plt.plot(petal_length, petal_width)
plt.title("Petal Length vs Petal Width")
plt.xlabel(column_names[2])
plt.ylabel(column_names[3])
plt.show()
```
