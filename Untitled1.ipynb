```python
import numpy as np
import matplotlib.pyplot as plt

# Read genome sequence from file
with open("genome.txt", "r") as file:
    genome_sequence = file.read().strip()  # Remove extra spaces/newlines

# Convert to list and calculate length
genome_list = list(genome_sequence)
genome_length = len(genome_list)

print(f"Genome Sequence Length: {genome_length}")

# Parametric equations for a helix
t = np.linspace(0, 4 * np.pi, genome_length)  # 4*pi gives about 2 turns
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)  # Spreads out the helix vertically


coordinates = np.column_stack((x, y, z))

color_map = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'yellow'}
colors = [color_map.get(base, 'black') for base in genome_list] 

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with colors
ax.scatter(x, y, z, c=colors, marker='o', s=50, alpha=0.9)  

# Labels and Title
ax.set_title("3D Helix Representation of Genome Sequence")
ax.set_xlabel("X ")
ax.set_ylabel("Y ")
ax.set_zlabel("Z (Height)")


ax.grid(True)
ax.set_xlim([-1.2, 1.2])
ax.set_ylim([-1.2, 1.2])
ax.set_zlim([0, 5])  # Same scale as given in the formula

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio.v2 as imageio

# Step 1: Load the image from a URL or local file
image_url = "https://upload.wikimedia.org/wikipedia/commons/4/47/PNG_transparency_demonstration_1.png"
image = imageio.imread(image_url)  
image_array = np.array(image)  


plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(image_array)
plt.title("Original Image")
plt.axis("off")

rot_img = np.rot90(image_array)
flip =np.fliplr(image_array)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(rot_img)
plt.title("Rotated Image")
plt.axis("off")


plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(flip)
plt.title("flip Image")
plt.axis("off")


#Y = 0.299*R + 0.587*G + 0.114*B
gray_img = np.dot (image_array[..., :3], [0.299, 0.587, 0.114])
plt.figure(figsize=(5, 5))
plt.imshow(gray_img, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petal width) #Accessing the target labels (species)
Y = np.array(iris.target)
feature_names = iris.feature_names

mean_values = np.mean(X, axis=0)
median_values = np.median(X, axis=0)
std_dev_values = np.std(X, axis=0)
min_values = np.min(X, axis=0)
max_values = np.max(X, axis=0)

print("Feature Names:", feature_names)
print("Mean Values:", mean_values)
print("Median Values:", median_values)
print("Standard Deviation:", std_dev_values)
print("Min Values:", min_values)
print("Max Values:", max_values)

sepal_features = X[:, :2] 

plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap="viridis", edgecolors='k')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Sepal Width")

# Histogram (Distribution of Sepal Length)

plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 2)
plt.hist(X[:, 0], bins=15, color="skyblue", edgecolor="black")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")

# Line Plot (Petal Length vs Petal Width)

plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 3)
plt.plot(X[:, 2], X[:, 3], 'bo-', markersize=3)  # Petal Length vs Petal Width
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs Petal Width")

plt.tight_layout()
plt.show()
```
