```python
import numpy as np
import matplotlib.pyplot as plt

group_A = np.array([12, 15, 14, 13, 16, 18, 19, 15, 14, 20, 17, 14, 15, 40, 45, 50, 62])
group_B = np.array([12, 17, 15, 13, 19, 20, 21, 18, 17, 16, 15, 14, 16, 15])

fig = plt.figure(figsize =(10, 7))

plt.boxplot(group_A)
plt.title('Group A BoxPlot')
plt.ylabel('Group A')
plt.show()

plt.boxplot(group_B)
plt.title('Group B BoxPlot')
plt.ylabel('Group B')
plt.show()
```

```python
from mpl_toolkits import mplot3d

f = open("genome.txt", "r")
data = f.read().strip()
f.close()

genome_sequence = list(data)
genome_length = len(genome_sequence)

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)
coordinates = np.column_stack((x, y, z))

# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

nucleotide = ["A", "T", "C", "G"]
colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
for i, nucleotide in enumerate(genome_sequence):
 ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
 color=colors[nucleotide], marker='o')

# Creating plot
ax.scatter3D(x, y, z)
plt.title("3D Scatter Plot")

# show plot
plt.show()
```

```python
from PIL import Image
from numpy import asarray

image = Image.open('img.jpeg')
numpydata = asarray(image)
plt.imshow(numpydata)
plt.show()

rotated_image= np.rot90(numpydata)
plt.imshow(rotated_image)
plt.show()

flipped_image = np.fliplr(numpydata)
plt.imshow(flipped_image)
plt.show()

gray_img = np.dot (numpydata[..., :3], [0.299, 0.587, 0.114])
plt.imshow(gray_img)
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) and target labels
X = np.array(iris.data)  # (Features (sepal length, sepal width, petal length, petal width) #Accessing the target labels (species)
Y = np.array(iris.target)  # Target variable (species: 0 for setosa, 1 for versicolor, 2 for virginica)

mean = np.mean(X, axis=0)
print("Mean of each feature:", mean)
median = np.median(X, axis=0)
print("Median of each feature:", median)
std = np.std(X, axis=0)
print("Standard deviation of each feature:", std)

min = np.min(X, axis=0)
print("Minimum values of each feature:", min)
max = np.max(X, axis=0)
print("Maximum values of each feature:", max)

sepal_features = X[:, :2]
print("Sepal length and width:\n", sepal_features)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis', edgecolor='k')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.colorbar(label='Species')
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=15, color='purple', edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram showing distribution of Sepal Length')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], 'go-', markersize=5)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Line Plot of Petal Length vs Petal Width')
plt.show()
```
