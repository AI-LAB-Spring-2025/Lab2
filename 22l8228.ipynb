```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


with open('genome.txt', 'r') as file:
    genome_sequence = file.read().replace('\n', '')

genome_list = list(genome_sequence)
genome_length = len(genome_list)

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)

# Assigning  colors
colors = {'A': 'red', 'T': 'green', 'C': 'blue', 'G': 'yellow'}
sc_arr = [colors[base] for base in genome_list]

# Creating 3D PLOT
visualize = plt.figure()
ax = visualize.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=sc_arr,s=180)


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Visualization')

plt.show()
```

```python
from PIL import Image
from numpy import asarray
import requests
from io import BytesIO
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


response = requests.get('https://i.natgeofe.com/n/548467d8-c5f1-4551-9f58-6817a8d2c45e/NationalGeographic_2572187_3x2.jpg')
img = Image.open(BytesIO(response.content))


numpydata = asarray(img)
plt.imshow(numpydata)
plt.show()


print(numpydata)

numpydata = np.rot90(numpydata, k = 2, axes = (0, 1))
plt.imshow(numpydata)
plt.show()

numpydata = np.fliplr(numpydata)
plt.imshow(numpydata)
plt.show()




gray_img = np.dot(numpydata[..., :3], [0.299, 0.587, 0.114])
plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
from PIL import Image
from numpy import asarray
import requests
from io import BytesIO
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petalwidth) #Accessing the target labels (species)
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 for versicolor, 2for virginica)
MEAN_VAL= np.mean(X,axis=0)
print(MEAN_VAL)
STD_VAL=np.std(X,axis=0)
print(STD_VAL)

Medians= np.median(X,axis=0)
print(Medians)
minimum = np.min(X, axis=0)
print(minimum)
maximum = np.max(X, axis=0)
print(maximum)

import matplotlib.pyplot as plt

# Scatter plot of sepal length vs sepal width
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Sepal Width")
plt.colorbar(label="Species")
plt.show()

# Histogram of sepal length
plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=20, color='skyblue', edgecolor='black')  # Adjust bins as needed
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()

# Line plot of petal length vs petal width
plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], color='coral', marker='o')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs Petal Width")
plt.show()



```
